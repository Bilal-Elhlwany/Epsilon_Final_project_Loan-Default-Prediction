## <a id="45">Visualize the data using proper visualization methods</a>

#### This function takes a pandas DataFrame as input and plots the distribution of each categorical column using pie charts. It first selects all columns with object data type, creates a color palette for the pie charts, and sets up subplots to display the charts in a grid. It then loops over each categorical column, gets the value counts for the column, sets up an explode list to separate the largest slice from the rest, and plots the pie chart with various styling options. Finally, it adds a grid to the plot and sets the title of the figure before displaying it.

```python
def Visual_categorical_Column(df):
    '''
    This function plots categorical variables as pie charts.
    '''
# Get a list of columns with categorical data
    cat_cols = df.select_dtypes(include=['object']).columns.to_list()
# Define color palette for pie charts
    colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#ffb3e6','#c2c2f0','#ffb3b3']   
# Create subplots with 7 columns per row
    fig, ax = plt.subplots((len(cat_cols) // 3) + 1, 3, figsize=(20, 25)) 
# Loop over categorical columns and plot each one as a pie chart
    for i, col in enumerate(cat_cols):
# Get value counts for the column
        counts = df[col].value_counts()
# Set up explode list to separate largest slice from the rest
        explode_list = [0] + [0.05] * (len(counts) - 1)
# Plot the pie chart
        counts.plot.pie(
            colors=colors,
            autopct='%0.0f%%',
            explode=explode_list,
            shadow=True,
            startangle=40,
            textprops={'color': 'black','fontsize':14},
            wedgeprops={'linewidth': 3, 'edgecolor':'grey'},
            rotatelabels='true',
            ax=ax[i // 3, i % 3],
        )  
# Set the title of the figure
    fig.suptitle('Column Distribution', y=0.98, ha='center', va='center', size=15, weight=150)
# Show the plot
    plt.show()
```

#### This function takes a pandas DataFrame and a column name as input, and plots the distribution of the target column using both a pie chart and a bar chart. It first sets up subplots for the two charts, and then plots the pie chart with various styling options. It then plots the bar chart using seaborn's countplot() function, and adds text labels to the bars. Finally, it sets the title of the figure and displays it.

```python
def Visual_Target_column(df, col):
    '''
    This function plots the target column with the distributions as both pie and bar charts.
    '''
# Set up subplots for pie chart and bar chart
    fig, ax = plt.subplots(1, 2, figsize=(20, 5))    
# Plot the pie chart
    df[col].value_counts().plot.pie(
        autopct='%.2f%%',
        explode=[0, 0.05],
        ax=ax[0],
        colors=['#ff9999', '#66b3ff'],
        shadow=True,
        startangle=40,
        textprops={'color': 'White', 'fontsize': 12},
        wedgeprops={'linewidth': 3, 'edgecolor': 'white'},
        rotatelabels='true',) 
# Plot the bar chart
    sns.countplot(
        data=df,
        x=col,
        ax=ax[1],
        palette='dark',
        edgecolor='white',
        linewidth=4, )    
# Add text labels to the bar chart
    ax[1].text(x=0, y=100000, s=df[col].value_counts()[0], color='white', va='center', ha='center')
    ax[1].text(x=1, y=25000, s=df[col].value_counts()[1], color='white', va='center', ha='center')
# Set the title of the figure
    fig.suptitle('Target Column Distribution', y=0.98, ha='center', va='center', size=15, weight=150)   
# Show the plot
    plt.show()
```

#### This function takes a pandas DataFrame as input and creates box plots for each numerical column in the DataFrame. It first selects all columns that are not of the object data type, calculates the number of rows required for the subplots, and creates the subplots. It then loops over each numerical column, plots a box plot for it using seaborn's boxplot() function, and sets the title and axis labels for the plot. Finally, it displays the plot.

```python
def box_plot_numerical_col(Num_data, figsize=(20, 40)):
    '''
    This function makes box plots for numerical columns in the given data frame.
    '''
# Get a list of numerical columns
    num_cols = [col for col in Num_data.columns if Num_data[col].dtype != 'object']
# Calculate the number of rows required for the subplots
    nrows = (len(num_cols) + 3) // 4
# Create the subplots
    fig, axes = plt.subplots(nrows, 4, figsize=figsize)
# Loop over each numerical column and plot a box plot for it
    for i, col in enumerate(num_cols):
        row = i // 4
        col = i % 4
        sns.boxplot(
            data=Num_data,
            y=num_cols[i],
            x='Status',
            ax=axes[row, col],
            width=0.5,
            linewidth=0.5,
            palette='dark')
        axes[row, col].set_ylabel('')
        axes[row, col].set_xlabel('')
        axes[row, col].set_title(col)
# Show the plot
    plt.show()
```

### To apply feature importance we need first :
    1.Replace null values in the dataset with mode 
    2.Encoding the categorical columns in the dataset 

* we use the fillna() method of the DataFrame to replace null values with the mode of each column, which is computed using the mode() method of the DataFrame.

* We use the iloc indexer to select the first row of the resulting mode DataFrame, which contains the mode of each column. Finally, we return the modified DataFrame.

```python
def replace_na(df):
    '''
    This function replaces null value in the dataset with the mode
    ''' 
    return df.fillna(df.mode().iloc[0]) # Replace null values with the mode of each column
```

* we use the select_dtypes() method of the DataFrame to select only the categorical columns, and create a list of their column names.

* We then loop through each categorical column, and use pandas' get_dummies() function to create a DataFrame of dummy variables for that column. We set the prefix parameter to the column name to distinguish the dummy variables, and set the drop_first parameter to True to avoid the dummy variable trap.

* We concatenate the dummy variable DataFrame with the original DataFrame using the concat() method, and drop the original categorical column using the drop() method.

```python
def dummify(df):
# Get the list of categorical columns
    cat_cols = df.select_dtypes(include='object').columns   
# Create dummy variables for each categorical column
    for col in cat_cols:
        dummies = pd.get_dummies(df[col], prefix=col, drop_first=True)
        df = pd.concat([df, dummies], axis=1)
        df.drop(col, axis=1, inplace=True)
# Return the modified DataFrame
    return df
```

## 1. Introduction

One of the most important data preprocessing step, we need to apply to our data is feature scaling. When we encounter any real world data set, the independent or feature variables may be mapped onto different scales.

This could cause some problems like feature with a higher value range make the model biased to this feature and it starts dominating .

we need to do scaling so that one significant number doesn’t impact the model just because of their large magnitude.

**Feature scaling** refers to procedures used to standardize these independent or feature variables so that they are mapped onto same scales.


Most of the ML algorithms perform well when the feature variables are mapped onto the same scale. They don’t perform well when features are mapped onto different scales. For example, in stochastic gradient descent, feature scaling can improve the convergence speed of the algorithm. In support vector machines, it can reduce the time to find support vectors. 


But, there are few exceptions as well. Decision trees and random forests are two of the algorithms where we don’t need to worry about feature scaling. These algorithms are scale invariant. Similarly, Naive Bayes and Linear Discriminant Analysis are not affected by feature scaling. In Short, any Algorithm which is not distance based is not affected by feature scaling.

<hr>

### There are 2 methods

1. Standardization with `MinMaxScaler`
2. Normalization with `StandardScaler`

**Advantages**<br>

- Normalization: shrinks the range such that the range is now between 0 and 1 (or -1 to 1 if there are negative values).
- Standardization: scales features such that the distribution is centered around 0, with a standard deviation of 1.


**Disadvantages**<br>

- Normalization: get influenced heavily by outliers (i.e. extreme values).(so we can't apply it )
- Standardization: not good if the data is not normally distributed (i.e. no Gaussian Distribution).





## 1.Normalizing data with MinMaxScaler


This technique of rescaling is also called **min-max scaling** or **min-max normalization**. **Normalization** refers to the rescaling of the features to a range of [0, 1], which is a special case of min-max scaling. So, in this technique, values are shifted and rescaled so that they end up ranging from zero to one. We do this by subtracting the minimum value (xmin ) and dividing by the maximum value (xmax ) minus the minimum value (xmin ). 


Mathematically, the new value x(i)norm of a sample x(i) can be calculated as follows:-

 
		     x(i)norm  =  (xi-  xmin )/(xmax-  xmin )
             
	
Here, x(i) is a particular sample value. xmax and xmin is the maximum and minimum feature value in a column.

	
Scikit-Learn provides a transformer called **MinMaxScaler** for this task. It has a feature range parameter to adjust the range of values. This estimator fit and transform each feature variable individually such that it is in the given range (between zero and one) on the training set. 

**MinMaxScaler** works well for the cases when the distribution is not normal or when the standard deviation is very small. But, it has one drawback. It is sensitive to outliers


The syntax for implementing min-max scaling procedure in Scikit-Learn is given as follows:- 


`from sklearn.preprocessing import MinMaxScaler`

`ms = MinMaxScaler()`

`X_train_ms = ms.fit_transform(X_train)`

`X_test_ms = ms.transform(X_test)`

## Define the function Need for  plotting the performance of the models 

```python
def plot_confmat_classreport(xTrain, yTrain, xTest, yTest, Class_Model_Name):
    #Calculating Prediction
    yPred = Class_Model_Name.predict(xTest)

    #Calculating Confusion Matrix and plotting it
    CM = confusion_matrix(yTest, yPred)
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
    sns.heatmap(CM, annot=True, fmt='g', cmap='Pastel1', ax=axes[0])
    axes[0].set_xlabel('Predicted')
    axes[0].set_ylabel('Truth')
    axes[0].set_title('Confusion Matrix')

    #Calculating Classification Report and plotting it
    classes = ['0', '1']
    visualizer = ClassificationReport(Class_Model_Name, classes=classes, support=True, ax=axes[1])
    visualizer.fit(xTrain, yTrain)
    visualizer.score(xTest, yTest)
    visualizer.show()
    axes[1].set_title('Classification Report')

    #Adjusting the layout of the subplots and displaying the plot
    fig.tight_layout()
    plt.show()

def plot_auc_accuracy(xTest, yTest, Class_Model_Name):
    #Calculating AUC and plotting the ROC Curve
    fpr, tpr, thresholds = roc_curve(yTest, Class_Model_Name.predict_proba(xTest)[:,1])
    roc_auc = roc_auc_score(yTest, Class_Model_Name.predict_proba(xTest)[:,1])
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
    axes[0].plot(fpr, tpr, color='purple', label='AUC = %0.2f' % roc_auc)
    axes[0].plot([0, 1], [0, 1], color='magenta', linestyle='--')
    axes[0].set_xlabel('False Positive Rate')
    axes[0].set_ylabel('True Positive Rate')
    axes[0].set_title('Receiver Operating Characteristic (ROC) Curve')
    
    #Calculating Accuracy and plotting it
    accuracy = Class_Model_Name.score(xTest, yTest)
    axes[1].bar(['Accuracy'], [accuracy], color=['purple'])
    axes[1].set_ylim([0, 1.5])
    axes[1].set_ylabel('Accuracy')
    axes[1].set_title('Accuracy')
    axes[1].set_xticks([])
    axes[1].tick_params(axis='y', labelsize=10)

    #Add accuracy value as text label on the accuracy bar plot
    axes[1].text(0, accuracy+0.05, '{:.2f}'.format(accuracy), horizontalalignment='center', fontsize=15)

    #Adjusting the layout of the subplots and displaying the plot
    fig.tight_layout()
    plt.show()
```
